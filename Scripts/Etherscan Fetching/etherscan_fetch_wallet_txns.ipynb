This notebook is used to fetch the first 2 tiers of token transfer transactions data from Etherscan involving Storj tokens. It starts from the list of node operators' wallets provided by Storj and grows to the next-tier receiver wallets.

```python
import pandas as pd
import requests
import time
import datetime as dt
import csv
import json
```

### Fetching 1st-tier Token Transfer Transactions

#### Data last fetched on June 19, 2021

```python
#Merge the 3 regional node data sets
regions = ['ap','eu','us']
files = []
for region in regions:
    files.append('data/{}_nodes_a.csv'.format(region))
```

```python
nodeOps = pd.DataFrame()
for file in files:
    df  = pd.read_csv(file)
    df['region'] = file[5:7]
    nodeOps = pd.concat([nodeOps, df])
    
```

```python
#Load up the list of wallets I've already fetched Etherscan data for
#Create a list of wallets I still need to fetch data for
toFetch = nodeOps['wallet'].unique()
```

```python
len(toFetch)
```

```python
t=0
error_wallets = []
```

```python
#for each node operator in the wallet list, fetch their token transfer txn history (sleeping for 5 seconds between each fetch)
with open('data/tokenTx_ori.csv', 'a', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    header = ['blockNumber','timeStamp','hash','nonce','blockHash','from','contractAddress','to','value',
                           'tokenName','tokenSymbol','tokenDecimal','transactionIndex','gas','gasPrice','gasUsed',
                           'cumulativeGasUsed','input','confirmations']
    writer.writerow(header)
    for wallet in toFetch:
        url = 'https://api.etherscan.io/api?module=account&action=tokentx&address={}&startblock=0&endblock=99999999&sort=asc&apikey=9JKAJTZZDRSX12NED3HED8E3TW9BC5X3ME'.format(wallet)
        mssg = 'OK'
        response = requests.get(url)
        jsonOP = response.json()
        parse2 = jsonOP['status']
        if parse2 == '0':
            mssg = jsonOP['message']
            if mssg == "No transactions found":
                pass
            else:
                error_wallets = error_wallets.append(wallet)
        else:
            parse1 = jsonOP['result']
            for i in parse1:
                writer.writerow(i.values())
        time.sleep(1.1)
        print(t, 'status: ',parse2, 'message: ',mssg)
        t+=1
        
```

```python
error_wallets
```

### Fetching Normal Transactions (Ignore This Part)

```python
tokentx = pd.read_csv('data/tokenTx_ori.csv')
```

```python
# Filtering to only token transfer transactions that used Storj token
tokentx_storj = tokentx[tokentx['tokenName']=='Storj']
len(tokentx_storj)
```

```python
# Number of txns that have value = 0
len(tokentx_storj[tokentx_storj['value']=='0'])
```

```python
# Number of txns where the sender and the receiver are the same wallet
len(tokentx_storj[tokentx_storj['from']==tokentx_storj['to']])
```

```python
# Number of wallets the were the receivers of Storj tokens in the 1st-tier token trf txns involving the node operators
receiver_wallets = tokentx[tokentx['tokenName']=='Storj']['to'].unique()
len(receiver_wallets)
```

```python
# Number of wallets the were the senders of Storj tokens in the 1st-tier token trf txns involving the node operators
sender_wallets = tokentx[tokentx['tokenName']=='Storj']['from'].unique()
len(sender_wallets)
```

```python
# Number of wallets the were either the senders or receivers of Storj tokens in the 1st-tier token trf txns involving the node operators
common_wallets = list(set(receiver_wallets) & set(sender_wallets))
len(common_wallets)
```

```python

other_sender_wallets = list(set(sender_wallets) - set(common_wallets))
len(other_sender_wallets)
```

```python
normalTx_receiver = pd.DataFrame(columns=['blockNumber','timeStamp','hash','nonce','blockHash','transactionIndex','from','to','value',
                           'gas','gasPrice','isError','txreceipt_status','input','contractAddress','cumulativeGasUsed','gasUsed',
                           'confirmations'])
normalTx_receiver.to_csv('data/normalTx_receiver.csv', index=False)

t=0
notx_wallets = []
error_wallets = []
```

### Combining First and Second Tiers

```python
tokenTx_ori = pd.read_csv('data/tokenTx_ori.csv')

string_cols = ['hash', 'blockHash', 'from', 'contractAddress', 'to', 'tokenName', 'tokenSymbol', 'input']
for col in string_cols:
    tokenTx_ori[col] = tokenTx_ori[col].str.lower()
    
len(tokenTx_ori)
```

```python
tokenTx_ori = tokenTx_ori[tokenTx_ori['tokenName'] == 'storj']
len(tokenTx_ori)
```

```python
tokenTx_ori = tokenTx_ori.drop_duplicates(subset=['blockNumber','timeStamp','hash','nonce','blockHash','from','contractAddress','to','value','tokenDecimal','transactionIndex','gas','gasPrice','gasUsed','cumulativeGasUsed','input'])
len(tokenTx_ori)
```

```python
tokenTx_receiver1 = pd.read_csv('data/tokenTx_receiver1.csv')

string_cols = ['hash', 'blockHash', 'from', 'contractAddress', 'to', 'tokenName', 'tokenSymbol', 'input']
for col in string_cols:
    tokenTx_receiver1[col] = tokenTx_receiver1[col].str.lower()
    
len(tokenTx_receiver1)
```

```python
tokenTx_receiver1 = tokenTx_receiver1[tokenTx_receiver1['tokenName'] == 'storj']
len(tokenTx_receiver1)
```

```python
tokenTx_receiver1 = tokenTx_receiver1.drop_duplicates(subset=['blockNumber','timeStamp','hash','nonce','blockHash','from','contractAddress','to','value','tokenDecimal','transactionIndex','gas','gasPrice','gasUsed','cumulativeGasUsed','input'])
len(tokenTx_receiver1)
```

```python
len(tokenTx_ori) + len(tokenTx_receiver1)
```

```python
tokenTx_2tiers = pd.concat([tokenTx_ori, tokenTx_receiver1])
len(tokenTx_2tiers)
```

```python
tokenTx_2tiers = tokenTx_2tiers.drop_duplicates(subset=['blockNumber','timeStamp','hash','nonce','blockHash','from','contractAddress','to','value','tokenDecimal','transactionIndex','gas','gasPrice','gasUsed','cumulativeGasUsed','input'])
len(tokenTx_2tiers)
```
